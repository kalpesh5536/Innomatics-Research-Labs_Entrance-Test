```python
# Create a temporary SQLite DB
import sqlite3

conn = sqlite3.connect("restaurants.db")
cursor = conn.cursor()

# Execute the SQL file
with open("restaurants.sql", "r") as file:
    sql_script = file.read()

cursor.executescript(sql_script)
conn.commit()

# Load SQL table into Pandas
restaurants = pd.read_sql("SELECT * FROM restaurants", conn)

# Run code
restaurants.head()
```

```python
orders_users = orders.merge(
    users,
    on="user_id",
    how="left"
)
```

```python
final_df = orders_users.merge(
    restaurants,
    on="restaurant_id",
    how="left"
)
```

# ALl MCQ

```python
import pandas as pd

df = pd.read_csv("final_food_delivery_dataset.csv")
df

results = {}
```

```python
# Q1) City with highest total revenue from Gold members

df[df["membership"]=="Gold"] \
  .groupby("city")["total_amount"] \
  .sum() \
  .sort_values(ascending=False)
```

```python
# Q.2) Which cuisine has the highest average order value across all orders?
df.groupby("cuisine")["total_amount"] \
  .mean() \
  .sort_values(ascending=False)

```

```python
# Q.3) How many distinct users spent more than ₹1000 in total?
(df.groupby("user_id")["total_amount"].sum() > 1000).sum()
```

```python
# Q.4) Which restaurant rating range generated the highest revenue?
import pandas as pd

bins = [0, 3.5, 4.0, 4.5, 5.0]
labels = ["3.0–3.5", "3.6–4.0", "4.1–4.5", "4.6–5.0"]

df["rating_range"] = pd.cut(df["rating"], bins=bins, labels=labels)

df.groupby("rating_range")["total_amount"] \
  .sum() \
  .sort_values(ascending=False)
```

```python
# Q.5) Among Gold members, which city has the highest average order value?
df[df["membership"] == "Gold"] \
  .groupby("city")["total_amount"] \
  .mean() \
  .sort_values(ascending=False)
```

```python
# Q.6) Cuisine with lowest restaurants but significant revenue
df.groupby("cuisine")["restaurant_id"].nunique(), \
df.groupby("cuisine")["total_amount"].sum()
```

```python
# Q.7) Percentage of total orders placed by Gold members
round(
    (df[df["membership"] == "Gold"].shape[0] / df.shape[0]) * 100
)
```

```python
# Q.8) Highest average order value but < 20 orders
df.groupby("restaurant_name_y") \
  .agg(
      avg_order=("total_amount", "mean"),
      orders=("order_id", "count")
  ) \
  .query("orders < 20") \
  .sort_values("avg_order", ascending=False) \
  .head()
```

```python
# Q.9) Which combination contributes highest revenue?
df.groupby(["membership", "cuisine"])["total_amount"] \
  .sum() \
  .sort_values(ascending=False)
```

```python
# 10) During which quarter is total revenue highest?
df["order_date"] = pd.to_datetime(df["order_date"], dayfirst=True)
df["quarter"] = df["order_date"].dt.to_period("Q")

df.groupby("quarter")["total_amount"] \
  .sum() \
  .sort_values(ascending=False)
```
